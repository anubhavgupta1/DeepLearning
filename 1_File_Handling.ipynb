```
with open('mobile_cleaned.csv','r') as file:
  print(file.readline())
```

```
# read entire file
with open('mobile_cleaned.csv','r') as file:
  print(file.read())
```

```
# read entire file line by line
with open('mobile_cleaned.csv','r') as file:
  for line in file:
    print(line)
```

```
# read entire file line by line
with open('mobile_cleaned.csv','r') as file:
  for line in file:
    for word in line.split(','):
      print(word)
    print('-----------------------')
```
